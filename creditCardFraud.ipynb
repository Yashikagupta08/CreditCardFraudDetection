#Pre-Processing

```python
#seperating features and labels
X = df.drop(columns = ['Class'], axis=1)
y = df['Class']
```

```python
#Using under sampling to decrease the legal transaction points to balance the dataset
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

#Class distribution before undersampling
Counter(y)
```

```python
#random under sampling
ros = RandomUnderSampler()
X, y = ros.fit_resample(X,y)

#Class distribution after undersampling
Counter(y)
```

```python
#new balanced dataframe 
df_new = pd.concat([X,y], axis=1)
```

```python
df_new
```

```python
#the difference is still clear after under sampling 
df_new.groupby('Class').mean()
```

```python
#To normalise the data features 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
#splitting the dataset in train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =2)
```

```python
#importing different metrics
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, classification_report, roc_curve, auc
```

```python
#function to calculate train and test accuracy
def accuracy(model, X_train, X_test, y_train, y_test):
  y_train_pred  = model.predict(X_train)
  train_accuracy = accuracy_score(y_train, y_train_pred)

  y_test_pred = model.predict(X_test)
  test_accuracy = accuracy_score(y_test, y_test_pred)

  return train_accuracy, test_accuracy
```

```python
#Plot the confusion matrix
def plotConfusionMatrix(model, X_test, y_test):
  cm=confusion_matrix(y_test, model.predict(X_test))
  plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Accent)
  category = ['Non-Fradulent', 'Fraudulent']
  plt.title('Confusion Matrix')
  plt.xlabel('Predicted Label')
  plt.ylabel('True Label')
  ticks = np.arange(len(category))
  plt.xticks(ticks, category)
  plt.yticks(ticks, category)
  s = [['TN', 'FP'], ['FN', 'TP']]

  for i in range(2):
    for j in range(2):
      plt.text(j,i, str(s[i][j]) + "=" +str(cm[i][j]), fontsize=12)

  plt.show()
```

#Logistic regression

```python
#Train the model
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter=1000)
model1.fit(X_train,y_train)
```

```python
#train and test accuracy after training
accuracy(model1, X_train, X_test, y_train, y_test)
```

```python
#confusion matrix
confusion_matrix(y_test, model1.predict(X_test))
```

```python
#plotting confusion matrix
plotConfusionMatrix(model1, X_test, y_test)
```

```python
#f1 score
f1_score(y_test, model1.predict(X_test), average='binary')
```

```python
#classification report
print(classification_report(y_test, model1.predict(X_test)))
```

```python
#function to plot roc curve
def plot_roc(model, X_text, y_test):
  predictions = model.predict_proba(np.array(X_test))
  fpr, tpr, threshold = roc_curve(y_test, predictions[:,1])
  roc_auc = auc(fpr, tpr)

  plt.title('ROC')
  plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)
  plt.legend(loc='lower right')
  plt.plot([0,1], [0,1], 'r--')
  plt.xlim([0,1])
  plt.ylim([0,1])
  plt.ylabel('True + Rate')
  plt.xlabel('False + Rate')
  plt.show()
```

```python
plot_roc(model1, X_test, y_test)
```
